### 权重衰减等价于L2范数正则化。

$$ \ell\left(w_{1}, w_{2}, b\right)=\frac{1}{n} \sum_{i=1}^{n} \frac{1}{2}\left(x_{1}^{(i)} w_{1}+x_{2}^{(i)} w_{2}+b-y^{(i)}\right)^{2} $$

$$ \ell\left(w_{1}, w_{2}, b\right)+\frac{\lambda}{2 n}\|\boldsymbol{w}\|^{2} $$

### 权重衰减的由来，如何理解权重衰减的含义。
看梯度下降公式：

$$ w_{1} \leftarrow\left(1-\frac{\eta \lambda}{|\mathcal{B}|}\right) w_{1}-\frac{\eta}{|\mathcal{B}|} \sum_{i \in \mathcal{B}} x_{1}^{(i)}\left(x_{1}^{(i)} w_{1}+x_{2}^{(i)} w_{2}+b-y^{(i)}\right) $$

L2范数正则化令权重w1先自乘小于1的数，再减去不含惩罚项的梯度。因此，L2范数正则化又叫做权重衰减。

权重衰减通过惩罚绝对值较大的模型参数为需要学习的模型增加了限制，这可能对过拟合有效。实际场景中，我们有时也在惩罚项中添加偏差元素的平方和

# 高维线性回归实验

以高维线性回归来引入一个过拟合问题：模型很复杂，但是样本数很少。

## 生成样本标签

```python
%matplotlib inline
import torch
import torch.nn as nn
import numpy as np
import sys
sys.path.append("..") 
import d2lzh_pytorch as d2l

n_train, n_test, num_inputs = 20, 100, 200
true_w, true_b = torch.ones(num_inputs, 1) * 0.01, 0.05

features = torch.randn((n_train + n_test, num_inputs))
# 线性函数
labels = torch.matmul(features, true_w) + true_b
# 线性函数再加上正态分布噪音
labels += torch.tensor(np.random.normal(0, 0.01, size=labels.size()), dtype=torch.float)
train_features, test_features = features[:n_train, :], features[n_train:, :]
train_labels, test_labels = labels[:n_train], labels[n_train:]
```

## 初始化模型参数

```python
def init_params():
    w = torch.randn((num_inputs,1),requires_grad = True)
    b = torch.zeros(1,requires_grad = True)
    return [w,b]
```

## 定义L2范数乘法项

```python
def l2_penalty(w):
    return (w**2).sum() / 2
```

## 定义训练和测试
在计算loss的时候，添加了L2范数惩罚项

```python
batch_size,num_epochs,lr = 1,100,0.003
net,loss = d2l.linreg, d2l.squared_loss

dataset = torch.utils.data.TensorDataset(train_features,train_labels)
train_iter = torch.utils.data.DataLoader(dataset=dataset,batch_size=batch_size,shuffle=True)

def fit_and_plot(lambd):
    w,b = init_params()
    train_ls, test_ls = [],[]
    for _ in range(num_epochs):
        for X,y in train_iter:
            # 添加了L2范数惩罚项
            l = loss(net(X,w,b),y) + lambd * l2_penalty(w)
#             print(l.shape)
            l = l.sum() # ?
            
            if w.grad is not None:
                w.grad.data.zero_()
                b.grad.data.zero_()
            l.backward()
            d2l.sgd([w,b],lr,batch_size)
            
        # 使用当前的模型在 测试集、 训练集 评估loss
        train_ls.append(loss(net(train_features,w,b),train_labels).mean().item())
        test_ls.append(loss(net(test_features,w,b),test_labels).mean().item())
    d2l.semilogy(range(1,num_epochs + 1), train_ls, 'epochs','loss',
                range(1,num_epochs + 1), test_ls,['train','test'])
    print('L2 norm of w:',w.norm().item())
    
        
        
```

## 观察过拟合

```python
fit_and_plot(lambd=0)
```

## 解决过拟合

```python
fit_and_plot(lambd=3)
```

# 简洁实现
optim的weight_decay参数

1.定义模型
2.设置参数的优化器
3.训练epoch
    3.1.按批次训练模型，前向传播、反向传播，更新模型（参数）
    3.2.使用更新过后的模型估计在训练集、测试集上的loss
4.绘制测试集/训练集上的loss-epoch曲线。

```python
%matplotlib inline
import torch
import torch.nn as nn
from torch.nn import init
import numpy as np
import sys
sys.path.append("..") 
import d2lzh_pytorch as d2l

n_train, n_test, num_inputs = 20, 100, 200
true_w, true_b = torch.ones(num_inputs, 1) * 0.01, 0.05

features = torch.randn((n_train + n_test, num_inputs))
# 线性函数
labels = torch.matmul(features, true_w) + true_b
# 线性函数再加上正态分布噪音
labels += torch.tensor(np.random.normal(0, 0.01, size=labels.size()), dtype=torch.float)
train_features, test_features = features[:n_train, :], features[n_train:, :]
train_labels, test_labels = labels[:n_train], labels[n_train:]
```

```python
batch_size,num_epochs,lr = 1,100,0.003
loss = d2l.squared_loss

dataset = torch.utils.data.TensorDataset(train_features,train_labels)
train_iter = torch.utils.data.DataLoader(dataset=dataset,batch_size=batch_size,shuffle=True)



def fit_and_plot_pytorch(wd):
    net = nn.Linear(num_inputs,1)
    
    init.normal_(net.weight, mean=0,std=1)
    init.normal_(net.bias,mean=0,std=1)
    
    optimizer_w = torch.optim.SGD(params=[net.weight],lr=lr,weight_decay=wd)# 对参数权重衰减
    optimizer_b = torch.optim.SGD(params=[net.bias],lr=lr) # 不对偏差进行参数衰减

    train_ls,test_ls = [],[]
    for _ in range(num_epochs):
        for X,y in train_iter:
            l = loss(net(X),y).mean()
            
            optimizer_w.zero_grad()
            optimizer_b.zero_grad()

            l.backward()

            # 对两个optimizer实例分别调用step 以更新参数
            optimizer_w.step()
            optimizer_b.step()
        
        train_ls.append(loss(net(train_features),train_labels).mean().item())
        test_ls.append(loss(net(test_features),test_labels).mean().item())
    
    d2l.semilogy(range(1, num_epochs + 1), train_ls, 'epochs', 'loss',
                 range(1, num_epochs + 1), test_ls, ['train', 'test'])
    print('L2 norm of w:', net.weight.data.norm().item())
    
```

```python
fit_and_plot_pytorch(0)
```
